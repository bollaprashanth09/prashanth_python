```python
# 📌 Weather Data Collector

import requests
import pandas as pd
from datetime import datetime

# Enter your OpenWeather API key
api_key ="cb9d57086955ec36e0faf8851d56037a"
 # your actual key here

# Ask user for city name
city = input("Enter city name: ")

# API Endpoint
url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

# Fetch data
response = requests.get(url)
data = response.json()

# Debug print
print("Raw Response:", data)

# Check for API errors
if data.get("cod") != 200:
    print("\n❌ Error:", data.get("message"))
else:
    # Extract required data
    temperature = data["main"]["temp"]
    humidity = data["main"]["humidity"]
    condition = data["weather"][0]["description"]
    time_collected = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    # Print Result
    print("\n✔ Weather Data Collected Successfully!")
    print(f"City: {city}")
    print(f"Temperature: {temperature}°C")
    print(f"Humidity: {humidity}%")
    print(f"Condition: {condition}")
    print(f"Time: {time_collected}")

    # Save to CSV
    df = pd.DataFrame([[city, temperature, humidity, condition, time_collected]],
                      columns=["City", "Temperature (°C)", "Humidity (%)", "Condition", "Time"])
    df.to_csv("weather_data.csv", mode='a', header=False, index=False)

    print("\n📁 Data saved to weather_data.csv")

    # Display table
    display(df)

```
